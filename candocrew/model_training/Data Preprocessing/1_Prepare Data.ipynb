## ReName File

```python
import shutil
import os
import glob

base_path = 'drive/My Drive/'
source_folder = os.path.join(base_path, 'PaymentReceipt','org_dataset')
desc_folder = os.path.join(base_path, 'PaymentReceipt','rename_dataset')

fileExtensions = ['.jpg', '.png', 'jpeg']

def rename_files(folderName, preFix):
  from_path = os.path.join(source_folder, folderName)
  to_path = os.path.join(desc_folder, folderName)

  # Create folder
  if not os.path.isdir(to_path):
          os.makedirs(to_path)

  files = []
  # Get filename list
  files = glob.glob(from_path + '/*')
  print(from_path)
  print(len(files))
  for index,fn in enumerate(files):
    if os.path.isfile(fn):
      # Get File Name and Extension
      file_name, file_ext = os.path.splitext(os.path.basename(fn))
      if file_ext.lower() in fileExtensions:
          old_file = os.path.join(from_path, file_name + file_ext)
          new_file = os.path.join(to_path, preFix + str(index) + file_ext)
          print('Org Name:',old_file)
          print('Rename:',new_file)
          os.rename(old_file, new_file)

# Rename & Move Payment Receipt Image
rename_files('KPay','KPay_')
rename_files('CBPay', 'CBPay_')
rename_files('AYAPay', 'AYAPay_')
rename_files('WavePay', 'WavePay_')
rename_files('Other', 'Other_')
```

## Spit Dataset

```python
import os
import shutil
import glob
import random
from sklearn.model_selection import train_test_split

def move_files(to_path, file_list):

    os.makedirs(to_path, exist_ok=True)

    for filePath in file_list:
         shutil.copy(filePath, to_path)

def split_train_test_val(name, sourceFolderPath, trainFolderPath, testFolderPath, valFolderPath):

    sourceDir =  os.path.join(sourceFolderPath, name)
    file_list = glob.glob(sourceDir + "/*")

    # train_temp=85%,test=15%
    train_temp_date_set_list,test_date_set_list=train_test_split(file_list, test_size=0.15, random_state=42)
    # train=85%,test=15% from train_temp
    train_date_set_list,val_date_set_list=train_test_split(train_temp_date_set_list,test_size=0.15, random_state=42)

    move_files(os.path.join(trainFolderPath, name), train_date_set_list)
    move_files(os.path.join(testFolderPath, name), test_date_set_list)
    move_files(os.path.join(valFolderPath, name), val_date_set_list)

base_path = 'drive/My Drive/'
sourceFolderPath = os.path.join(base_path, 'PaymentReceipt','rename_dataset')
train_dir = os.path.join(base_path, 'PaymentReceipt', 'dataset/train')
test_dir = os.path.join(base_path, 'PaymentReceipt', 'dataset/test')
val_dir= os.path.join(base_path, 'PaymentReceipt', 'dataset/val')

# Create train,val,test dataset folder
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Split train,test,validation dataset
split_train_test_val('KPay', sourceFolderPath, train_dir, test_dir, val_dir)
split_train_test_val('AYAPay', sourceFolderPath, train_dir, test_dir, val_dir)
split_train_test_val('CBPay',sourceFolderPath, train_dir, test_dir, val_dir)
split_train_test_val('WavePay', sourceFolderPath, train_dir, test_dir, val_dir)
split_train_test_val('Other', sourceFolderPath, train_dir, test_dir, val_dir)
```

## Count the Number of Files

```python
import glob
import os


def count_files_in_directory(directory_path):
    fileCount = 0
    for f in os.listdir(directory_path):
        file_list = glob.glob(os.path.join(directory_path, f) + "/*")
        fileCount += len(file_list)
    return fileCount

rawdataset = os.path.join(base_path, 'PaymentReceipt','raw_dataset')
train_dir = os.path.join(base_path, 'PaymentReceipt', 'dataset/train')
test_dir = os.path.join(base_path, 'PaymentReceipt', 'dataset/test')
val_dir= os.path.join(base_path, 'PaymentReceipt', 'dataset/val')

print(f"Number of files in the raw dataset: {count_files_in_directory(rawdataset)}")
print(f"Number of files in the train dataset: {count_files_in_directory(train_dir)}")
print(f"Number of files in the test dataset: {count_files_in_directory(test_dir)}")
print(f"Number of files in the validation dataset: {count_files_in_directory(val_dir)}")
```
